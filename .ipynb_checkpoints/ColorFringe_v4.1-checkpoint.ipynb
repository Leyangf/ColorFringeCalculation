## 2. Data Fitting and Common Sampling Scheme

```python
# CHL fit: polynomial of degree 6

# x_CHL contains the wavelength values and y_CHL the corresponding CHL values from the dataset CHLzf85.
x_CHL = CHLzf85[:, 0]
y_CHL = CHLzf85[:, 1]

# Fit a 6th-degree polynomial to the CHL data
CHL_coeffs = np.polyfit(x_CHL, y_CHL, 6)

# Define a function to evaluate the fitted polynomial at any given x (wavelength)
def CHLFit(x):
    # np.polyval evaluates the polynomial for the input x using coefficients in descending order
    return np.polyval(CHL_coeffs, x)

# Optional: Print the coefficients for inspection
print("Fitted CHL coefficients:", CHL_coeffs)

x_fit = np.linspace(400, 700, 300)
CHL_fit = CHLFit(x_fit)
print("CHL_fit shape:", CHL_fit.shape)
```

```python
# Light Source Fit: polynomial of degree 6 and resampling

def resample_spectrum_poly(spectrum, new_wavelengths, degree=6):
    # Separate the wavelength and intensity data
    wavelengths = spectrum[:, 0]
    intensities = spectrum[:, 1]
    
    # Fit a polynomial of the given degree to the data
    coeffs = np.polyfit(wavelengths, intensities, degree)
    poly_func = np.poly1d(coeffs)
    
    # Evaluate the polynomial at the new wavelengths
    new_intensities = poly_func(new_wavelengths)

    # Normalize the intensities so that the maximum is 100
    max_intensity = np.max(new_intensities)
    new_intensities = new_intensities / max_intensity * 100
    
    return np.column_stack((new_wavelengths, new_intensities))

# Resample the spectrum using a 6th-degree polynomial fit
Daylight = resample_spectrum_poly(or_Daylight, x_CHL, degree=6)

# Print the resampled spectrum
print("Resampled Daylight Spectrum:", Daylight.shape)
```

## 3. CHL Curve Modification and Visualization

```python
import numpy as np
from scipy.interpolate import interp1d, UnivariateSpline

# -------------------------------
# Pre-calculations and Setup
# -------------------------------

# Compute the wavelength interval (dx) from the fitted wavelength array.
dx = x_fit[1] - x_fit[0]

# Find the original lowest point on the CHL curve within the 400â700 nm range.
# np.argmin returns the index of the minimum value in the CHL_fit array.
x_min = x_fit[np.argmin(CHL_fit)]
y_min = np.min(CHL_fit)
print("Original lowest point at x = {:.2f} nm, y = {:.2f}".format(x_min, y_min))

# Create interpolators for the CHL curve:
# 1. Linear interpolator for the right section.
linear_fit = interp1d(x_fit, CHL_fit, kind='linear', fill_value="extrapolate")
# 2. Spline (UnivariateSpline) for the left section with no smoothing (s=0).
spline_fit = UnivariateSpline(x_fit, CHL_fit, s=0)

# -------------------------------
# Modified CHL Curve Function
# -------------------------------
def CHLFit_mod_overall(x_vals, x_shift=0.0, tilt=0.0):
    """
    Compute the modified CHL curve by applying a tilt on either side
    of a pivot point. The modification involves:
    
    - A horizontal shift of the pivot point.
    - Alteration of the derivative (tilt) factor:
        * On the left side of the pivot (x < pivot): the derivative is scaled by (1 - tilt)
        * On the right side (x >= pivot): the derivative is scaled by (1 + tilt)
    
    The function then reconstructs the CHL curve by integrating these modified derivatives.
    
    Parameters:
        x_vals (array-like): Wavelength values (assumed to be sorted and equally spaced).
        x_shift (float): Horizontal shift applied to the pivot point.
        tilt (float): Tilt factor to modulate the slope on each side of the pivot.
    
    Returns:
        np.ndarray: The modified CHL curve values corresponding to the given wavelengths.
    """
    # Determine spacing from x_vals; use a default of 1 if array length is 1.
    dx = x_vals[1] - x_vals[0] if len(x_vals) > 1 else 1

    # Define the pivot as the location near the minimum (x_min) with an applied horizontal shift.
    pivot_x = x_min + x_shift  
    # Evaluate the original CHL curve at the pivot location using an external function CHLFit.
    # (Assumes CHLFit is defined elsewhere in your code base.)
    pivot_y = CHLFit(x_min)
    
    # Apply horizontal shift to all x values.
    x_shifted = x_vals - x_shift
    
    # Evaluate base curve values and compute derivatives using the two interpolators.
    # Left section: using the spline-based interpolation and its derivative.
    base_left = spline_fit(x_shifted)
    deriv_left = spline_fit.derivative()(x_shifted)
    
    # Right section: using linear interpolation.
    base_right = linear_fit(x_shifted)
    # Calculate the derivative numerically using np.gradient.
    deriv_right = np.gradient(base_right, dx)
    
    # Select the appropriate derivative for each wavelength based on the pivot.
    # If x < pivot_x, use the spline derivative; else use the linear derivative.
    base_deriv = np.where(x_vals < pivot_x, deriv_left, deriv_right)
    
    # Apply different tilt factors for the two regions:
    # Left side uses (1 - tilt), right side uses (1 + tilt).
    new_deriv = np.where(x_vals < pivot_x, base_deriv * (1 - tilt), base_deriv * (1 + tilt))
    
    # Initialize an array to hold the modified CHL values.
    g = np.empty_like(x_vals)
    # Identify the index of the pivot point (the x value closest to pivot_x).
    pivot_index = np.argmin(np.abs(x_vals - pivot_x))
    g[pivot_index] = pivot_y  # Set the pivot value to be the original pivot_y.

    # -------------------------------
    # Reconstruct the Modified CHL Curve by Integration
    # -------------------------------

    # Forward integration for wavelengths greater than or equal to pivot.
    if pivot_index < len(x_vals) - 1:
        # Cumulative sum of derivatives times spacing approximates integration.
        g[pivot_index + 1:] = pivot_y + np.cumsum(new_deriv[pivot_index:-1]) * dx

    # Backward integration for wavelengths less than the pivot.
    if pivot_index > 0:
        # Integrate in reverse order and reverse the result.
        g[:pivot_index] = pivot_y - np.cumsum(new_deriv[1:pivot_index + 1][::-1])[::-1] * dx

    return g

# -------------------------------
# Sampling Function for Modified Values
# -------------------------------
def extract_sampled_modified_values_overall(x_shift=0.0, tilt=0.0):
    """
    Sample the modified CHL curve at fixed intervals (every 10 nm) within the wavelength range 400â700 nm.
    
    Parameters:
        x_shift (float): Horizontal shift applied to the pivot in the modified CHL function.
        tilt (float): Tilt factor applied to modify the derivative in the CHL function.
    
    Returns:
        np.ndarray: A two-column matrix. First column contains the sampled wavelengths and 
                    the second column contains the corresponding floored CHL values.
    """
    # Create an array of wavelengths from 400 nm to 700 nm in steps of 10 nm.
    x_sample = np.arange(400, 701, 10)
    # Compute the modified CHL curve values at these wavelengths.
    y_sample = CHLFit_mod_overall(x_sample, x_shift, tilt)
    # Floor the values to get integer CHL values.
    y_sample_int = np.floor(y_sample)
    # Combine the wavelength and integer CHL values into a 2-column matrix.
    return np.column_stack((x_sample, y_sample_int))
```

```python
# Import necessary libraries.
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import FloatSlider, Output, interact
from IPython.display import display

# -------------------------------
# Global Variable
# -------------------------------
# Global variable to store the sampled modified CHL values.
global_CHLfit = None

# -------------------------------
# Functions to Update and Retrieve the Global CHL Array
# -------------------------------
def get_modified_CHL_array(x_shift=0.0, tilt=0.0):
    """
    Update and return the global global_CHLfit array based on the provided x_shift and tilt.
    
    Parameters:
        x_shift (float): Horizontal shift applied to the CHL pivot.
        tilt (float): Tilt factor applied to the CHL curve.
        
    Returns:
        np.ndarray: The updated global_CHLfit array containing the sampled modified CHL values.
    """
    global global_CHLfit
    # Call the previously defined sampling function to update the global array.
    global_CHLfit = extract_sampled_modified_values_overall(x_shift, tilt)
    return global_CHLfit
    

# -------------------------------
# Function to Plot the Modified CHL Curve
# -------------------------------
def plot_CHL_overall(x_shift=0.0, tilt=0.0):
    """
    Plot the modified CHL curve along with the original data points and pivot markers.
    
    The plot includes:
        - The modified CHL curve (using the CHLFit_mod_overall function).
        - Original data points (plotted in red).
        - The original lowest point on the CHL curve (in green).
        - The shifted pivot point marker (in blue).
    
    Parameters:
        x_shift (float): Horizontal shift to apply at the pivot point.
        tilt (float): Tilt factor to modify the CHL curve.
    """
    # Compute the modified CHL values for the full wavelength fit range.
    y_vals = CHLFit_mod_overall(x_fit, x_shift, tilt)
    
    # Create a new figure and plot the modified CHL curve.
    plt.figure(figsize=(8, 4))
    plt.plot(x_fit, y_vals, label='Modified CHL Curve (Opposite Tilts)', lw=2)
    # Overlay the original data points.
    plt.scatter(x_CHL, y_CHL, color='red', label='Data Points')
    # Mark the original lowest CHL point.
    plt.scatter([x_min], [y_min], color='green', s=100, label='Original Lowest Point')
    # Mark the shifted pivot point.
    plt.scatter([x_min + x_shift], [CHLFit(x_min)], color='blue', s=100, label='Shifted Pivot')
    
    # Set labels, title, and grid for clarity.
    plt.xlabel('Wavelength (nm)')
    plt.ylabel('CHL Value')
    plt.title('Interactive CHL Curve with Opposite Tilts')
    plt.legend()
    plt.grid(True)
    plt.xlim(400, 700)
    plt.ylim(-100, 700)
    plt.show()

# -------------------------------
# Set Up Interactive Widgets for Real-Time Adjustments
# -------------------------------

# Create sliders for x_shift and tilt with defined ranges and steps.
x_shift_slider = FloatSlider(min=-80, max=80, step=1, value=0, description='x Shift (nm)')
tilt_slider = FloatSlider(min=-0.8, max=0.8, step=0.01, value=0, description='Tilt Factor')

# Create an Output widget to capture and display plot outputs.
out = Output()

def update_display(x_shift, tilt):
    """
    Update the plot and global_CHLfit array based on slider inputs.
    
    This function is called automatically whenever the interactive widget values change.
    
    Parameters:
        x_shift (float): New x_shift value from the slider.
        tilt (float): New tilt factor from the slider.
    """
    with out:
        out.clear_output(wait=True)
        # Plot the modified CHL curve with current parameter values.
        plot_CHL_overall(x_shift, tilt)
        # Update the global CHL array using the defined function.
        updated_array = get_modified_CHL_array(x_shift, tilt)
        # Print the updated global_CHLfit array for inspection.
        print("Updated global_CHLfit array:")
        print(updated_array)

# Create the interactive interface with the update_display function as a callback.
interact(update_display, x_shift=x_shift_slider, tilt=tilt_slider)
display(out)
```

```python
def wavelength_to_rgb(wavelength, gamma=1):
    """
    Convert a wavelength in nanometers to an approximate RGB color.
    
    This function maps a visible wavelength (approximately 380 nm to 780 nm)
    to an RGB color using an algorithm adapted from Dan Bruton:
    http://www.efg2.com/Lab/ScienceAndEngineering/Spectra.htm

    It calculates the color components based on which segment of the visible
    spectrum the wavelength falls into. Gamma correction is applied to adjust 
    the intensity for display.

    Parameters:
        wavelength (float): The wavelength in nanometers.
        gamma (float, optional): The gamma correction factor (default is 1).

    Returns:
        tuple: A tuple (R, G, B) where each value is a float in the range [0, 1].
               Wavelengths outside the 380 to 780 nm range return black (0, 0, 0).
    """
    # If the wavelength is in the violet segment (380-440 nm)
    if 380 <= wavelength < 440:
        # Calculate an attenuation factor to decrease intensity at the range boundaries.
        attenuation = 0.3 + 0.7 * (wavelength - 380) / (440 - 380)
        # Red increases as wavelength decreases from 440 to 380 nm.
        R = ((-(wavelength - 440) / (440 - 380)) * attenuation) ** gamma
        # Green is essentially absent in this range.
        G = 0.0
        # Blue is at full intensity, modulated by the attenuation.
        B = (1.0 * attenuation) ** gamma

    # For the blue to cyan transition (440-490 nm)
    elif 440 <= wavelength < 490:
        R = 0.0
        # Green increases from 0 to full intensity as wavelength increases.
        G = ((wavelength - 440) / (490 - 440)) ** gamma
        # Blue remains at full intensity.
        B = 1.0 ** gamma

    # For the cyan to green transition (490-510 nm)
    elif 490 <= wavelength < 510:
        R = 0.0
        # Green holds at full intensity.
        G = 1.0 ** gamma
        # Blue decreases linearly from full intensity to 0.
        B = (-(wavelength - 510) / (510 - 490)) ** gamma

    # For the green to yellow transition (510-580 nm)
    elif 510 <= wavelength < 580:
        # Red increases from 0 to full as wavelength goes from 510 to 580 nm.
        R = ((wavelength - 510) / (580 - 510)) ** gamma
        # Green remains at full intensity.
        G = 1.0 ** gamma
        B = 0.0

    # For the yellow to red transition (580-645 nm)
    elif 580 <= wavelength < 645:
        # Red remains at full intensity.
        R = 1.0 ** gamma
        # Green decreases linearly from full to 0.
        G = (-(wavelength - 645) / (645 - 580)) ** gamma
        B = 0.0

    # For the red segment (645-780 nm)
    elif 645 <= wavelength <= 780:
        # Attenuation decreases as wavelength increases, reducing brightness.
        attenuation = 0.3 + 0.7 * (780 - wavelength) / (780 - 645)
        R = (1.0 * attenuation) ** gamma
        G = 0.0
        B = 0.0

    # For wavelengths outside the visible range, return black.
    else:
        R = G = B = 0.0

    return (R, G, B)
```

```python
# %% Cell 2: Sensor Data Processing, Edge Functions & Grid Search

# -----------------------------------------------
# Data Setup for Sensor Channels
# -----------------------------------------------
# Convert sensor data from the normalized sensor arrays into NumPy arrays
# with float64 precision to ensure computational accuracy.
SensorBluedata = np.asarray(SensorBlue_norm[:, 1], dtype=np.float64)
SensorGreendata = np.asarray(SensorGreen_norm[:, 1], dtype=np.float64)
SensorReddata  = np.asarray(SensorRed_norm[:, 1], dtype=np.float64)

# -----------------------------------------------
# Global Constants and Parameters
# -----------------------------------------------
# K: A constant used in the sensor data processing or model.
K = 1.4

# xrange_val: The range for x values in plots (e.g., setting limits for visualization).
xrange_val = 200

# defocusrange: A parameter that likely controls the defocus range in your analysis.
defocusrange = 1000

# tol: Tolerance value for the grid search or convergence criteria.
tol = 0.15

# F_value: A parameter that may be used in model fitting or sensor data processing.
F_value = 8

# g_value: A parameter likely related to the sensor processing or edge function computation.
g_value = 10

# K_value: Although equal to 'K', this might be used in a separate context or for clarity in different formulas.
K_value = 1.4
```

```python
# --- JIT-Compiled Functions for PSF and Exposure ---

# Note: Ensure you have the required imports from Numba and math.
from numba import njit
from math import sqrt, fabs, erf as math_erf

@njit
def Exposure_jit(x, F):
    """
    Compute the exposure function using a hyperbolic tangent scaling.

    The hyperbolic tangent function is normalized by its maximum value (tanh(F)).
    This function scales an input 'x' using parameter 'F'.

    Parameters:
        x (float): Input value (can be an averaged weighted sensor value).
        F (float): Exposure parameter that controls the steepness.

    Returns:
        float: Normalized exposure value.
    """
    return np.tanh(F * x) / np.tanh(F)

@njit
def linear_PSF_jit(x: float, ratio: float) -> float:
    """
    Compute a linear point spread function (PSF) for a given x value and ratio.

    The PSF gives a weight that transitions linearly between 0 and 1 based on 'ratio'.
    For very small 'ratio', the function returns a step-function-like behavior.

    Parameters:
        x (float): The x-coordinate (or offset).
        ratio (float): Scaling factor, determining the linear region.

    Returns:
        float: Weight computed using a linear scaling based on the ratio.
    """
    if ratio < 1e-6:
        # If the ratio is nearly zero, act like a step function.
        return 1.0 if x >= 0 else 0.0
    if x >= ratio:
        return 1.0
    elif x <= -ratio:
        return 0.0
    else:
        # Linearly interpolate between 0 and 1.
        return 0.5 * (1 + x / ratio)

@njit
def gaussian_PSF_jit(x: float, ratio: float) -> float:
    """
    Compute a Gaussian-based point spread function (PSF) for a given x value and ratio.

    This function uses the error function to generate a smooth transition in weights,
    representing the Gaussian PSF. For very small 'ratio', it returns a step-function-like behavior.

    Parameters:
        x (float): The x-coordinate (or offset).
        ratio (float): Scaling factor, determining the width of the Gaussian.

    Returns:
        float: Weight computed using a Gaussian function approximation.
    """
    if ratio < 1e-6:
        return 1.0 if x >= 0 else 0.0
    # Use the error function to generate a smooth transition.
    return 0.5 * (1.0 + math_erf(x / (sqrt(2) * ratio)))

@njit
def compute_edge_jit(x: float, z: float, F: float, g: float,
                     sensor_data, CHLdata, dx: float, K: float,
                     psf_mode: str) -> float:
    """
    Compute the edge response by integrating sensor responses weighted by a PSF.

    For each data point in the CHL dataset, the corresponding sensor data is weighted
    by a PSF (either linear or Gaussian). The weighted sum (numerator) is normalized by the
    total sensor sum (denominator) and then processed through the exposure function
    (with an exponentiation by g) to produce the final edge value.

    Parameters:
        x (float): The x-coordinate for which the edge is computed.
        z (float): A reference position (typically related to CHL data).
        F (float): Exposure parameter influencing the tanh function in Exposure_jit.
        g (float): Gamma correction factor applied after exposure computation.
        sensor_data (array-like): Array of sensor channel data.
        CHLdata (array-like): Array of CHL data points.
        dx (float): Spatial increment (used if numerical integration over space is required).
        K (float): Constant used to compute the denominator factor (affects PSF scaling).
        psf_mode (str): Mode for point spread function; 'gaussian' or 'linear'.

    Returns:
        float: The computed edge response after exposure and gamma correction.
    """
    # Calculate a denominator factor based on the constant K.
    denom_factor = sqrt(4 * K**2 - 1)
    
    # Initialize the numerator that will accumulate the weighted sensor values.
    num = 0.0
    for n in range(len(CHLdata)):
        # Determine the ratio for weighting based on the difference between z and each CHL data point.
        ratio = fabs((z - CHLdata[n]) / denom_factor)
        # Choose the PSF function based on the selected mode (Gaussian or linear).
        if psf_mode.lower() == "linear":
            weight = linear_PSF_jit(x, ratio)
        else:
            weight = gaussian_PSF_jit(x, ratio)
        # Accumulate the weighted sensor data.
        num += sensor_data[n] * weight

    # Sum up the sensor data for normalization.
    den = 0.0
    for n in range(len(sensor_data)):
        den += sensor_data[n]
    
    # Compute the exposure value, apply gamma correction, and return the final edge response.
    return Exposure_jit(num / den, F) ** g
```

```python
# --- Edge Functions Wrappers ---
def EdgeR_gaussian(x: float, z: float, F: float, g: float,
                   CHLdata, dx: float, K: float, psf_mode: str = "gaussian") -> float:
    """Edge function for the red sensor channel (Gaussian PSF)."""
    return compute_edge_jit(x, z, F, g, SensorReddata, CHLdata, dx, K, psf_mode)


def EdgeG_gaussian(x: float, z: float, F: float, g: float,
                   CHLdata, dx: float, K: float, psf_mode: str = "gaussian") -> float:
    """Edge function for the green sensor channel (Gaussian PSF)."""
    return compute_edge_jit(x, z, F, g, SensorGreendata, CHLdata, dx, K, psf_mode)


def EdgeB_gaussian(x: float, z: float, F: float, g: float,
                   CHLdata, dx: float, K: float, psf_mode: str = "gaussian") -> float:
    """Edge function for the blue sensor channel (Gaussian PSF)."""
    return compute_edge_jit(x, z, F, g, SensorBluedata, CHLdata, dx, K, psf_mode)


def EdgeR_linear(x: float, z: float, F: float, g: float,
                 CHLdata, dx: float, K: float, psf_mode: str = "gaussian") -> float:
    """Edge function for the red sensor channel (Linear PSF)."""
    return compute_edge_jit(x, z, F, g, SensorReddata, CHLdata, dx, K, psf_mode)


def EdgeG_linear(x: float, z: float, F: float, g: float,
                 CHLdata, dx: float, K: float, psf_mode: str = "gaussian") -> float:
    """Edge function for the green sensor channel (Linear PSF)."""
    return compute_edge_jit(x, z, F, g, SensorGreendata, CHLdata, dx, K, psf_mode)


def EdgeB_linear(x: float, z: float, F: float, g: float,
                 CHLdata, dx: float, K: float, psf_mode: str = "gaussian") -> float:
    """Edge function for the blue sensor channel (Linear PSF)."""
    return compute_edge_jit(x, z, F, g, SensorBluedata, CHLdata, dx, K, psf_mode)
```

```python
# Use CHLdata_0 from the global modified CHL array.
# (Ensure that global_CHLfit has been updated via get_modified_CHL_array before using.)
CHLdata_0 = global_CHLfit[:, 1] if global_CHLfit is not None else None

# --- Helper Functions for Color Fringe Calculations ---
def Farbsaum(x, z, F, g, use_ideal, CHLdata, psf_mode: str = "gaussian"):
    """
    Determine if a significant color difference exists (binary decision)
    using sensor edge responses.

    Returns:
        int: 1 if the absolute differences between channels exceed tol, else 0.
    """
    if use_ideal:
        r = EdgeR_linear(x, z, F, g, CHLdata, dx, K, psf_mode)
        g_val = EdgeG_linear(x, z, F, g, CHLdata, dx, K, psf_mode)
        b = EdgeB_linear(x, z, F, g, CHLdata, dx, K, psf_mode)
    else:
        r = EdgeR_gaussian(x, z, F, g, CHLdata, dx, K, psf_mode)
        g_val = EdgeG_gaussian(x, z, F, g, CHLdata, dx, K, psf_mode)
        b = EdgeB_gaussian(x, z, F, g, CHLdata, dx, K, psf_mode)
    return 1 if (abs(r - b) > tol or abs(r - g_val) > tol or abs(g_val - b) > tol) else 0


def Farbsaumbreite(z, F, g, use_ideal, CHLdata, psf_mode: str = "gaussian"):
    """
    Compute the overall color fringe width (binary method) for a given defocus (z).

    Returns:
        int: Aggregated binary width value.
    """
    xs = np.arange(-int(np.floor(xrange_val)), int(np.floor(xrange_val)) + 1)
    return np.sum([Farbsaum(x, z, F, g, use_ideal, CHLdata, psf_mode) for x in xs])


def ColorFringe(x, z, F, g, use_ideal, psf_mode: str = "gaussian"):
    """
    Compute the color fringe responses for a given x position.

    Returns:
        Tuple: (EdgeR, EdgeG, EdgeB) values.
    """
    if use_ideal:
        return (EdgeR_linear(x, z, F, g, CHLdata_0, dx, K, psf_mode),
                EdgeG_linear(x, z, F, g, CHLdata_0, dx, K, psf_mode),
                EdgeB_linear(x, z, F, g, CHLdata_0, dx, K, psf_mode))
    else:
        return (EdgeR_gaussian(x, z, F, g, CHLdata_0, dx, K, psf_mode),
                EdgeG_gaussian(x, z, F, g, CHLdata_0, dx, K, psf_mode),
                EdgeB_gaussian(x, z, F, g, CHLdata_0, dx, K, psf_mode))
```

```python
def evaluate_params(params):
    """
    Evaluate a single (x_shift, tilt) parameter combination by computing the 
    aggregated farbsaumbreite value over the range of defocus positions.

    For each (x_shift, tilt):
      - Compute the modified CHL curve and extract CHL data.
      - For each defocus position (z) in z_vals, calculate the Farbsaumbreite.
      - Aggregate these values using the selected aggregation_mode (mean or max).
        For the "mean" mode, only the range between the left and right global maximum
        fringe values (based on their positions in the array) is averaged.

    Parameters:
        params (tuple): (x_shift, tilt)

    Returns:
        tuple: (x_shift, tilt, aggregated_width)
    """
    x_shift, tilt = params
    # Compute the modified CHL curve for the current parameters.
    CHL_matrix = extract_sampled_modified_values_overall(x_shift, tilt)
    CHLdata_local = CHL_matrix[:, 1]
    
    # Calculate Farbsaumbreite for each defocus position.
    fringe_values = [
        Farbsaumbreite(z, F_value, 1, use_ideal=False, CHLdata=CHLdata_local, psf_mode="gaussian")
        for z in z_vals
    ]
    
    # Aggregate the fringe values based on the selected aggregation_mode.
    if aggregation_mode == "mean":
        # Split the fringe values at the midpoint.
        mid_index = len(fringe_values) // 2
        
        # Find the index of the maximum value in the left segment (from the start to the midpoint).
        left_segment = fringe_values[:mid_index+1]  # include midpoint for safety
        left_max_idx = np.argmax(left_segment)
        
        # Find the index of the maximum value in the right segment (from the midpoint to the end).
        right_segment = fringe_values[mid_index:]
        right_max_idx = np.argmax(right_segment) + mid_index  # adjust index relative to whole array
        
        # Use the smaller index as left and the larger as right (in case the order is swapped).
        left_index = min(left_max_idx, right_max_idx)
        right_index = max(left_max_idx, right_max_idx)
        
        # Compute the mean only over the values between (and including) the two global maximum indices.
        aggregated_width = np.mean(fringe_values[left_index:right_index+1])
    
    elif aggregation_mode == "max":
        aggregated_width = np.max(fringe_values)
    else:
        aggregated_width = np.mean(fringe_values)  # default fallback
    
    return (x_shift, tilt, aggregated_width)


def grid_search():
    """
    Perform a grid search over x_shift and tilt parameters using parallel threads.

    Returns:
        tuple: (best_params, results_array, x_shift_values, tilt_values)
    """
    # Define grid ranges for x_shift and tilt.
    x_shift_values = np.linspace(-80, 80, num=21)
    tilt_values = np.linspace(-0.8, 0.8, num=21)
    param_grid = [(x_shift, tilt) for x_shift in x_shift_values for tilt in tilt_values]
    
    # Evaluate each parameter combination in parallel.
    with concurrent.futures.ThreadPoolExecutor() as executor:
        results = list(executor.map(evaluate_params, param_grid))
    
    results_array = np.array(results)
    # Find the combination with the smallest aggregated fringe width.
    min_index = np.argmin(results_array[:, 2])
    best_params = results_array[min_index]
    return best_params, results_array, x_shift_values, tilt_values
```
