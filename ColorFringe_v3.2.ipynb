## 2. Data Fitting and Common Sampling Scheme

```python
# CHL fit: polynomial of degree 6

# x_CHL contains the wavelength values and y_CHL the corresponding CHL values from the dataset CHLzf85.
x_CHL = CHLzf85[:, 0]
y_CHL = CHLzf85[:, 1]

# Fit a 6th-degree polynomial to the CHL data
CHL_coeffs = np.polyfit(x_CHL, y_CHL, 6)

# Define a function to evaluate the fitted polynomial at any given x (wavelength)
def CHLFit(x):
    # np.polyval evaluates the polynomial for the input x using coefficients in descending order
    return np.polyval(CHL_coeffs, x)

# Optional: Print the coefficients for inspection
print("Fitted CHL coefficients:", CHL_coeffs)

x_fit = np.linspace(400, 700, 300)
CHL_fit = CHLFit(x_fit)
print("CHL_fit shape:", CHL_fit.shape)
```

```python
# Light Source Fit: polynomial of degree 6 and resampling

def resample_spectrum_poly(spectrum, new_wavelengths, degree=6):
    # Separate the wavelength and intensity data
    wavelengths = spectrum[:, 0]
    intensities = spectrum[:, 1]
    
    # Fit a polynomial of the given degree to the data
    coeffs = np.polyfit(wavelengths, intensities, degree)
    poly_func = np.poly1d(coeffs)
    
    # Evaluate the polynomial at the new wavelengths
    new_intensities = poly_func(new_wavelengths)

    # Normalize the intensities so that the maximum is 100
    max_intensity = np.max(new_intensities)
    new_intensities = new_intensities / max_intensity * 100
    
    return np.column_stack((new_wavelengths, new_intensities))

# Resample the spectrum using a 6th-degree polynomial fit
Daylight = resample_spectrum_poly(or_Daylight, x_CHL, degree=6)

# Print the resampled spectrum
print("Resampled Daylight Spectrum:", Daylight.shape)
```

## 3. CHL Curve Modification and Visualization

```python
# %% Cell 1: Core Functions & Global Array Access
# Make sure that x_fit, CHLFit, x_CHL, y_CHL, and CHL_fit are defined in your environment.

# Compute the wavelength interval (dx)
dx = x_fit[1] - x_fit[0]

# Find the original lowest point on the CHL curve (in the 400â700 nm range)
x_min = x_fit[np.argmin(CHL_fit)]
y_min = np.min(CHL_fit)
print("Original lowest point at x = {:.2f} nm, y = {:.2f}".format(x_min, y_min))

def CHLFit_mod_overall(x_vals, x_shift=0.0, tilt=0.0):
    """
    Returns the modified CHL curve with opposite tilts applied to each side of the pivot.
    Uses vectorized integration via np.cumsum.
    """
    pivot_x = x_min + x_shift
    pivot_y = CHLFit(x_min)
    
    # Compute the horizontally shifted base curve.
    x_shifted = x_vals - x_shift
    base = CHLFit(x_shifted)
    
    # Compute numerical derivative.
    dbase = np.gradient(base, dx)
    
    # Apply opposite tilt factors using vectorized np.where.
    new_deriv = np.where(x_vals < pivot_x, dbase * (1 - tilt), dbase * (1 + tilt))
    
    # Integrate the new derivative:
    g = np.empty_like(x_vals)
    # Find pivot index.
    pivot_index = np.argmin(np.abs(x_vals - pivot_x))
    g[pivot_index] = pivot_y
    
    # Forward integration from the pivot.
    if pivot_index < len(x_vals) - 1:
        g[pivot_index+1:] = pivot_y + np.cumsum(new_deriv[pivot_index:-1]) * dx
    # Backward integration from the pivot.
    if pivot_index > 0:
        # Reverse the segment, take cumulative sum, then reverse back.
        g[:pivot_index] = pivot_y - np.cumsum(new_deriv[1:pivot_index+1][::-1])[::-1] * dx
    
    return g

def extract_sampled_modified_values_overall(x_shift=0.0, tilt=0.0):
    """
    Samples the modified CHL curve (with opposite tilts) every 10 nm between 400 and 700 nm.
    Returns a 2-column matrix: [wavelength, integer CHL value].
    """
    x_sample = np.arange(400, 701, 10)
    y_sample = CHLFit_mod_overall(x_sample, x_shift, tilt)
    y_sample_int = np.floor(y_sample)
    return np.column_stack((x_sample, y_sample_int))
```

```python
# Global variable to store the sampled modified CHL values.
global_CHLfit = None

def get_modified_CHL_array(x_shift=0.0, tilt=0.0):
    """
    Computes and returns the global_CHLfit array given the input x_shift and tilt.
    This function can be called anywhere in your notebook.
    """
    global global_CHLfit
    global_CHLfit = extract_sampled_modified_values_overall(x_shift, tilt)
    return global_CHLfit

# -------------------------------------------
import matplotlib.pyplot as plt
from ipywidgets import FloatSlider, interact, Output

def plot_CHL_overall(x_shift=0.0, tilt=0.0):
    """
    Plots the modified CHL curve with opposite tilts on each side of the pivot.
    """
    y_vals = CHLFit_mod_overall(x_fit, x_shift, tilt)
    
    plt.figure(figsize=(8, 4))
    plt.plot(x_fit, y_vals, label='Modified CHL Curve (Opposite Tilts)', lw=2)
    plt.scatter(x_CHL, y_CHL, color='red', label='Data Points')
    plt.scatter([x_min], [y_min], color='green', s=100, label='Original Lowest Point')
    plt.scatter([x_min + x_shift], [CHLFit(x_min)], color='blue', s=100, label='Shifted Pivot')
    
    plt.xlabel('Wavelength (nm)')
    plt.ylabel('CHL Value')
    plt.title('Interactive CHL Curve with Opposite Tilts')
    plt.legend()
    plt.grid(True)
    plt.xlim(400, 700)
    plt.show()

# Define interactive sliders.
x_shift_slider = FloatSlider(min=-50, max=50, step=1, value=0, description='x Shift (nm)')
tilt_slider = FloatSlider(min=-0.5, max=0.5, step=0.01, value=0, description='Tilt Factor')

# Output area for interactive updates.
out = Output()

def update_display(x_shift, tilt):
    with out:
        out.clear_output(wait=True)
        # Display the interactive plot.
        plot_CHL_overall(x_shift, tilt)
        # Update and display the global_CHLfit array.
        updated_array = get_modified_CHL_array(x_shift, tilt)
        print("Updated global_CHLfit array:")
        print(updated_array)

# Create the interactive interface.
interact(update_display, x_shift=x_shift_slider, tilt=tilt_slider)
display(out)
```

```python
# Function converts a wavelength (in nanometers) from the visible spectrum into an approximate RGB color value.

def wavelength_to_rgb(wavelength, gamma=0.8):
    """
    Convert a given wavelength in nm to an RGB color.
    Based on the algorithm from Dan Bruton:
    http://www.efg2.com/Lab/ScienceAndEngineering/Spectra.htm
    """
    if 380 <= wavelength < 440:
        attenuation = 0.3 + 0.7*(wavelength - 380) / (440 - 380)
        R = ((-(wavelength - 440) / (440 - 380)) * attenuation) ** gamma
        G = 0.0
        B = (1.0 * attenuation) ** gamma
    elif 440 <= wavelength < 490:
        R = 0.0
        G = ((wavelength - 440) / (490 - 440)) ** gamma
        B = 1.0 ** gamma
    elif 490 <= wavelength < 510:
        R = 0.0
        G = 1.0 ** gamma
        B = (-(wavelength - 510) / (510 - 490)) ** gamma
    elif 510 <= wavelength < 580:
        R = ((wavelength - 510) / (580 - 510)) ** gamma
        G = 1.0 ** gamma
        B = 0.0
    elif 580 <= wavelength < 645:
        R = 1.0 ** gamma
        G = (-(wavelength - 645) / (645 - 580)) ** gamma
        B = 0.0
    elif 645 <= wavelength <= 780:
        attenuation = 0.3 + 0.7*(780 - wavelength) / (780 - 645)
        R = (1.0 * attenuation) ** gamma
        G = 0.0
        B = 0.0
    else:
        R = G = B = 0.0

    return (R, G, B)
```

```python
# %% Cell 2: Imports, Data Setup, and Global Parameters
SensorBluedata = np.asarray(SensorBlue_norm[:, 1], dtype=np.float64)
SensorGreendata = np.asarray(SensorGreen_norm[:, 1], dtype=np.float64)
SensorReddata  = np.asarray(SensorRed_norm[:, 1], dtype=np.float64)

# Global constants and parameters
K = 1.4
xrange_val = 200    # Range for x values in plots
defocusrange = 1000
tol = 0.05
F_value = 8
g_value = 10
K_value = 1.4
```

```python
# =============================================================================
# Core Calculation Functions (JIT-compiled with Numba)
# =============================================================================
# %% JIT-Compiled Functions
@njit
def Exposure_jit(x, F):
    # Note: np.tanh is supported in numba.
    return np.tanh(F * x) / np.tanh(F)

# JIT-compiled Linear PSF function
@njit
def linear_PSF_jit(x: float, ratio: float) -> float:
    if ratio < 1e-6:
        return 1.0 if x >= 0 else 0.0
    if x >= ratio:
        return 1.0
    elif x <= -ratio:
        return 0.0
    else:
        return 0.5 * (1 + x / ratio)

# JIT-compiled Gaussian PSF function
@njit
def gaussian_PSF_jit(x: float, ratio: float) -> float:
    if ratio < 1e-6:
        return 1.0 if x >= 0 else 0.0
    return 0.5 * (1.0 + math_erf(x / (sqrt(2) * ratio)))

@njit
def compute_edge_jit(x: float, z: float, F: float, g: float,
                     sensor_data, CHLdata, dx: float, K: float) -> float:
    """
    Compute the edge response using sensor and CHL data.

    Parameters:
        x (float): x-coordinate.
        z (float): Reference position.
        F (float): Exposure parameter.
        g (float): Gamma correction factor.
        sensor_data (array-like): Data from a sensor channel.
        CHLdata (array-like): CHL data.
        dx (float): Spatial increment.
        K (float): Constant for denominator factor.

    Returns:
        float: Computed edge value.
    """
    denom_factor = sqrt(4 * K**2 - 1)
    num = 0.0
    for n in range(len(CHLdata)):
        ratio = fabs((z - CHLdata[n]) / denom_factor)
        weight = gaussian_PSF_jit(x, ratio)
        num += sensor_data[n] * weight
    den = 0.0
    for n in range(len(sensor_data)):
        den += sensor_data[n]
    return Exposure_jit(num / den, F) ** g

# Wrappers for sensor channels (ensure Sensor*data globals are defined)
def EdgeR_jit(x: float, z: float, F: float, g: float,
              CHLdata, dx: float, K: float) -> float:
    """Edge function for the red sensor channel."""
    return compute_edge_jit(x, z, F, g, SensorReddata, CHLdata, dx, K)

def EdgeG_jit(x: float, z: float, F: float, g: float,
              CHLdata, dx: float, K: float) -> float:
    """Edge function for the green sensor channel."""
    return compute_edge_jit(x, z, F, g, SensorGreendata, CHLdata, dx, K)

def EdgeB_jit(x: float, z: float, F: float, g: float,
              CHLdata, dx: float, K: float) -> float:
    """Edge function for the blue sensor channel."""
    return compute_edge_jit(x, z, F, g, SensorBluedata, CHLdata, dx, K)
```

```python
# Aliases for interactive selection.
ideal_weight = linear_PSF_jit
gaussian_weight = gaussian_PSF_jit

# Define a range of x values and a ratio parameter.
x = np.linspace(-100, 100, 400)
ratio = 20  # adjust as needed

# Compute weights using the two functions.
y_ideal = np.array([ideal_weight(val, ratio) for val in x])
y_gaussian = np.array([gaussian_weight(val, ratio) for val in x])

# Define the hard ideal step function.
y_step = np.where(x < 0, 0, 1)

# Plot all three functions.
plt.figure(figsize=(8, 4))
plt.plot(x, y_gaussian, label=r'Gaussian Weight: $0.5(1+\mathrm{erf}(x/(\sqrt{2}\,ratio)))$')
plt.plot(x, y_ideal, label='Ideal Weight (with linear interpolation)', linestyle='--')
plt.plot(x, y_step, label='Ideal Step Function', linestyle=':')
plt.xlabel("x")
plt.ylabel("Weight")
plt.title("Comparison of Weighting Functions")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Use CHLdata_0 from global_CHLfit
CHLdata_0 = global_CHLfit[:, 1] 

def EdgeR_linear(x: float, z: float, F: float, g: float,
                 CHLdata, dx: float, K: float) -> float:
    return compute_edge_jit(x, z, F, g, SensorReddata, CHLdata, dx, K)

def EdgeG_linear(x: float, z: float, F: float, g: float,
                 CHLdata, dx: float, K: float) -> float:
    return compute_edge_jit(x, z, F, g, SensorGreendata, CHLdata, dx, K)

def EdgeB_linear(x: float, z: float, F: float, g: float,
                 CHLdata, dx: float, K: float) -> float:
    return compute_edge_jit(x, z, F, g, SensorBluedata, CHLdata, dx, K)

# --- Helper Functions for Binary and Color Fringe Calculations ---
def Farbsaum(x, z, F, g, use_ideal, CHLdata):
    if use_ideal:
        r = EdgeR_linear(x, z, F, g, CHLdata, dx, K)
        g_val = EdgeG_linear(x, z, F, g, CHLdata, dx, K)
        b = EdgeB_linear(x, z, F, g, CHLdata, dx, K)
    else:
        r = EdgeR_jit(x, z, F, g, CHLdata, dx, K)
        g_val = EdgeG_jit(x, z, F, g, CHLdata, dx, K)
        b = EdgeB_jit(x, z, F, g, CHLdata, dx, K)
    return 1 if (abs(r - b) > tol or abs(r - g_val) > tol or abs(g_val - b) > tol) else 0

def Farbsaumbreite(z, F, g, use_ideal, CHLdata):
    xs = np.arange(-int(np.floor(xrange_val)), int(np.floor(xrange_val)) + 1)
    return np.sum([Farbsaum(x, z, F, g, use_ideal, CHLdata) for x in xs])

def ColorFringe(x, z, F, g, use_ideal):
    if use_ideal:
        return (EdgeR_linear(x, z, F, g, CHLdata_0, dx, K),
                EdgeG_linear(x, z, F, g, CHLdata_0, dx, K),
                EdgeB_linear(x, z, F, g, CHLdata_0, dx, K))
    else:
        return (EdgeR_jit(x, z, F, g, CHLdata_0, dx, K),
                EdgeG_jit(x, z, F, g, CHLdata_0, dx, K),
                EdgeB_jit(x, z, F, g, CHLdata_0, dx, K))

# --- Interactive Plotting Code (Spot Diagram Removed) ---
def update_plots(z, F, gamma, model):
    use_ideal = True if model == 'Ideal Weight' else False

    # Create a figure with a GridSpec layout: two plots side by side.
    fig = plt.figure(figsize=(12, 6))
    gs = fig.add_gridspec(1, 2)
    
    # Left: Edge spread functions.
    ax1 = fig.add_subplot(gs[0, 0])
    xs = np.linspace(-xrange_val, xrange_val, 400)
    if use_ideal:
        edgeR_vals = np.array([EdgeR_linear(x, z, F, gamma, CHLdata_0, dx, K) for x in xs])
        edgeG_vals = np.array([EdgeG_linear(x, z, F, gamma, CHLdata_0, dx, K) for x in xs])
        edgeB_vals = np.array([EdgeB_linear(x, z, F, gamma, CHLdata_0, dx, K) for x in xs])
    else:
        edgeR_vals = np.array([EdgeR_jit(x, z, F, gamma, CHLdata_0, dx, K) for x in xs])
        edgeG_vals = np.array([EdgeG_jit(x, z, F, gamma, CHLdata_0, dx, K) for x in xs])
        edgeB_vals = np.array([EdgeB_jit(x, z, F, gamma, CHLdata_0, dx, K) for x in xs])
    farb_vals = np.array([Farbsaum(x, z, F, gamma, use_ideal, CHLdata_0) for x in xs])
    
    ax1.plot(xs, edgeR_vals, color=(1, 0, 0), label='EdgeR')
    ax1.plot(xs, edgeG_vals, color=(0, 1, 0), label='EdgeG')
    ax1.plot(xs, edgeB_vals, color=(0, 0, 1), label='EdgeB')
    ax1.plot(xs, farb_vals, color=(0.2, 0.2, 0.2), label='Farbsaum')
    ax1.set_title(f"Edge Functions ({model} Weighting)")
    ax1.set_xlabel("x")
    ax1.set_ylabel("Response")
    ax1.set_ylim(0, 1)
    ax1.legend()
    ax1.grid(True)
    
    # Right: Density plot using ColorFringe (1D slice replicated vertically)
    ax2 = fig.add_subplot(gs[0, 1])
    X = np.linspace(-xrange_val, xrange_val, 200)
    Y = np.linspace(0, 0.5 * xrange_val, 100)
    XX, YY = np.meshgrid(X, Y)
    
    img = np.zeros((YY.shape[0], YY.shape[1], 3))
    for i in range(XX.shape[0]):
        for j in range(XX.shape[1]):
            img[i, j, :] = ColorFringe(XX[i, j], z, F, gamma, use_ideal)
    img = np.clip(img, 0, 1)
    ax2.imshow(img, extent=[-xrange_val, xrange_val, 0, 0.5 * xrange_val], aspect='auto')
    ax2.set_title("Density Plot (ColorFringe)")
    ax2.set_xlabel("x")
    ax2.set_ylabel("y")
    
    plt.tight_layout()
    plt.show()
    
    # Compute and print overall color width using the binary method.
    width = Farbsaumbreite(z, F, gamma, use_ideal, CHLdata_0)
    print("Color width (Binary method) =", width)

# --- Interactive Widget ---
interact(update_plots,
         z=FloatSlider(min=-defocusrange, max=defocusrange, step=1, value=0, description="Defocus"),
         F=FloatSlider(min=1, max=16, step=1, value=F_value, description="Exposure F"),
         gamma=FloatSlider(min=0.5, max=20, step=0.5, value=g_value, description="Gamma"),
         model=RadioButtons(options=['Ideal Weight', 'Gaussian Weight'], description='Model'))
```

```python
dd = 10  # step size for defocus values
z_vals = np.arange(-defocusrange, defocusrange + 1, dd)  # defocus positions along the optical axis

def plot_farbs(F_value, label, hue, use_ideal):
    """
    For each z in `z_vals`, calculate the "Farbsaumbreite" (color fringe width)
    using the binary method and plot these widths versus the defocus values.
    
    Parameters:
        F_value (float): The exposure parameter.
        label (str): Label for the plot.
        hue (float): Hue value for the plot color (used with plt.cm.hsv).
        use_ideal (bool): If True, use the ideal (linear) weighting; else, use Gaussian.
    """
    # Here g is set to 1.
    data = np.array([[z, Farbsaumbreite(z, F_value, g_value, use_ideal, CHLdata_0)] for z in z_vals])
    plt.plot(data[:, 0], data[:, 1], 'o-', markersize=3, label=label, color=plt.cm.hsv(hue))
    return data

def plot_farbs_comparison(F_values=[2], use_ideal=True):
    """
    Plot Farbsaumbreite for various F values using the selected weighting method.
    
    Parameters:
        F_values (list): List of exposure (F) values to plot. Defaults to [2].
        use_ideal (bool): If True, use the ideal (linear) weighting; if False, use Gaussian.
    """
    plt.figure(figsize=(8, 4))
    
    # Generate hues based on the number of F_values.
    hues = np.linspace(0.0, 0.75, len(F_values))
    data_dict = {}
    
    for F_val, hue in zip(F_values, hues):
        data_dict[F_val] = plot_farbs(F_val, f"F={F_val}", hue, use_ideal)
    
    model_name = "Ideal Weight" if use_ideal else "Gaussian Weight"
    plt.title(f"Farbsaumbreite for various F values (g=1) using {model_name} (Binary method)")
    plt.xlabel("z (defocus)")
    plt.ylabel("Farbsaumbreite")
    plt.legend()
    plt.grid(True)
    plt.show()
    return data_dict

def interactive_plot(F_values, weight_function_choice):
    # Map the selection to a Boolean flag.
    use_ideal = True if weight_function_choice == 'Ideal Weight' else False
    # Ensure F_values is a sorted list.
    F_values = sorted(F_values)
    plot_farbs_comparison(F_values=F_values, use_ideal=use_ideal)

# Create interactive widgets:
F_values_widget = widgets.SelectMultiple(
    options=[1, 2, 4, 8, 16, 32],
    value=[8],
    description='F values',
    disabled=False
)

weight_function_widget = widgets.RadioButtons(
    options=['Ideal Weight', 'Gaussian Weight'],
    value='Ideal Weight',
    description='Model',
    disabled=False
)

# Display interactive controls:
interact(interactive_plot, 
         F_values=F_values_widget, 
         weight_function_choice=weight_function_widget);
```

```python
# =============================================================================
# Parameter Evaluation and Grid Search
# =============================================================================
def evaluate_params(params):
    """
    Evaluate a single (x_shift, tilt) parameter combination.
    
    For each combination:
      - Compute the modified CHL curve and extract local CHL data.
      - Calculate the binary fringe width using the helper function Farbsaum.
        Aggregation is performed according to the global variable aggregation_mode.
    
    Returns:
        Tuple (x_shift, tilt, aggregated_width)
    """
    x_shift, tilt = params
    # extract_sampled_modified_values_overall must be defined elsewhere.
    CHL_matrix = extract_sampled_modified_values_overall(x_shift, tilt)
    
    CHLdata_local = CHL_matrix[:, 1]
    
    xs = np.arange(-int(np.floor(xrange_val)), int(np.floor(xrange_val)) + 1)
    fringe_values = [Farbsaum(x, 0.0, F_value, 1, use_ideal=False, CHLdata=CHLdata_local) for x in xs]
    
    if aggregation_mode == "max":
        aggregated_width = max(fringe_values)
    elif aggregation_mode == "mean":
        aggregated_width = sum(fringe_values) / len(fringe_values)
    else:
        aggregated_width = max(fringe_values)
    
    return (x_shift, tilt, aggregated_width)

def grid_search():
    """
    Perform a grid search over x_shift and tilt parameters using parallel threads.
    
    Returns:
        best_params: The parameter combination (x_shift, tilt, aggregated_width) with the smallest width.
        results_array: All evaluated parameter combinations as a numpy array.
        x_shift_values: The grid of x_shift values.
        tilt_values: The grid of tilt values.
    """
    x_shift_values = np.linspace(-50, 50, num=21)
    tilt_values = np.linspace(-0.5, 0.5, num=21)
    param_grid = [(x_shift, tilt) for x_shift in x_shift_values for tilt in tilt_values]
    
    with concurrent.futures.ThreadPoolExecutor() as executor:
        results = list(executor.map(evaluate_params, param_grid))
    
    results_array = np.array(results)
    min_index = np.argmin(results_array[:, 2])
    best_params = results_array[min_index]
    return best_params, results_array, x_shift_values, tilt_values
```

```python
# =============================================================================
# Main Execution Block
# =============================================================================
aggregation_mode = "mean"  # or "max"
if __name__ == "__main__":
    # Run the grid search.
    start_time = time.time()
    best_params, results_array, x_shift_values, tilt_values = grid_search()
    end_time = time.time()
    
    print("Best parameters found based on", aggregation_mode, "width:")
    print("x_shift: {:.2f}, tilt: {:.2f}, width_value: {:.2f}".format(*best_params))
    print("Grid search took {:.2f} seconds".format(end_time - start_time))
    
    # Plotting the aggregated width values.
    plot_start_time = time.perf_counter()
    
    # Reshape the aggregated width values into a 2D array.
    width_matrix = results_array[:, 2].reshape((len(x_shift_values), len(tilt_values)))
    
    # Transpose the matrix so that x_shift is on the horizontal axis.
    plt.figure(figsize=(8, 6))
    plt.imshow(width_matrix.T,
               extent=[x_shift_values.min(), x_shift_values.max(),
                       tilt_values.min(), tilt_values.max()],
               origin='lower', aspect='auto')
    plt.xlabel('x_shift')
    plt.ylabel('tilt')
    plt.title(f'Color Fringe Width ({aggregation_mode}) vs. x_shift and tilt')
    plt.colorbar(label=f'width_{aggregation_mode}')
    plt.show()
    
    plot_end_time = time.perf_counter()
    print(f"Plotting cell run time: {plot_end_time - plot_start_time:.4f} seconds")
```
