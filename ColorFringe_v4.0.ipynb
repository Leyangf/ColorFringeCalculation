## 2. Data Fitting and Common Sampling Scheme

```python
# CHL fit: polynomial of degree 6

# x_CHL contains the wavelength values and y_CHL the corresponding CHL values from the dataset CHLzf85.
x_CHL = CHLzf85[:, 0]
y_CHL = CHLzf85[:, 1]

# Fit a 6th-degree polynomial to the CHL data
CHL_coeffs = np.polyfit(x_CHL, y_CHL, 6)

# Define a function to evaluate the fitted polynomial at any given x (wavelength)
def CHLFit(x):
    # np.polyval evaluates the polynomial for the input x using coefficients in descending order
    return np.polyval(CHL_coeffs, x)

# Optional: Print the coefficients for inspection
print("Fitted CHL coefficients:", CHL_coeffs)

x_fit = np.linspace(400, 700, 300)
CHL_fit = CHLFit(x_fit)
print("CHL_fit shape:", CHL_fit.shape)
```

```python
# Light Source Fit: polynomial of degree 6 and resampling

def resample_spectrum_poly(spectrum, new_wavelengths, degree=6):
    # Separate the wavelength and intensity data
    wavelengths = spectrum[:, 0]
    intensities = spectrum[:, 1]
    
    # Fit a polynomial of the given degree to the data
    coeffs = np.polyfit(wavelengths, intensities, degree)
    poly_func = np.poly1d(coeffs)
    
    # Evaluate the polynomial at the new wavelengths
    new_intensities = poly_func(new_wavelengths)

    # Normalize the intensities so that the maximum is 100
    max_intensity = np.max(new_intensities)
    new_intensities = new_intensities / max_intensity * 100
    
    return np.column_stack((new_wavelengths, new_intensities))

# Resample the spectrum using a 6th-degree polynomial fit
Daylight = resample_spectrum_poly(or_Daylight, x_CHL, degree=6)

# Print the resampled spectrum
print("Resampled Daylight Spectrum:", Daylight.shape)
```

## 3. CHL Curve Modification and Visualization

```python
# Compute the wavelength interval (dx)
dx = x_fit[1] - x_fit[0]

# Find the original lowest point on the CHL curve (in the 400â700 nm range)
x_min = x_fit[np.argmin(CHL_fit)]
y_min = np.min(CHL_fit)
print("Original lowest point at x = {:.2f} nm, y = {:.2f}".format(x_min, y_min))
```

```python
from scipy.interpolate import UnivariateSpline, interp1d
import numpy as np

# Linear interpolator for the right section.
linear_fit = interp1d(x_fit, CHL_fit, kind='linear', fill_value="extrapolate")
# Spline for the left section.
spline_fit = UnivariateSpline(x_fit, CHL_fit, s=0)

def CHLFit_mod_overall(x_vals, x_shift=0.0, tilt=0.0):
    """
    Compute the modified CHL curve by applying an opposite tilt on either side 
    of a pivot point. The left section (x < pivot) uses a spline-based interpolation 
    and derivative, while the right section (x >= pivot) uses linear interpolation.

    Parameters:
        x_vals (array-like): Wavelength values.
        x_shift (float): Horizontal shift applied to the pivot.
        tilt (float): Tilt factor.

    Returns:
        np.ndarray: Modified CHL curve values.
    """
    # Determine spacing from x_vals; assume x_vals is sorted and equally spaced.
    dx = x_vals[1] - x_vals[0] if len(x_vals) > 1 else 1
    
    # Define pivot. x_min should be defined externally (e.g., 400).
    pivot_x = x_min + x_shift  
    pivot_y = CHLFit(x_min)  # Use the original curve function at x_min for the pivot value.
    
    # Shift wavelengths.
    x_shifted = x_vals - x_shift
    
    # Evaluate base values and derivatives for each section.
    # Left section (using spline)
    base_left = spline_fit(x_shifted)
    deriv_left = spline_fit.derivative()(x_shifted)
    
    # Right section (using linear interpolation)
    base_right = linear_fit(x_shifted)
    deriv_right = np.gradient(base_right, dx)
    
    # Combine the derivative values based on the position relative to pivot_x.
    base_deriv = np.where(x_vals < pivot_x, deriv_left, deriv_right)
    
    # Apply tilt factors: (1 - tilt) on the left side and (1 + tilt) on the right side.
    new_deriv = np.where(x_vals < pivot_x, base_deriv * (1 - tilt), base_deriv * (1 + tilt))
    
    # Integrate the modified derivative to reconstruct the modified CHL curve.
    g = np.empty_like(x_vals)
    pivot_index = np.argmin(np.abs(x_vals - pivot_x))
    g[pivot_index] = pivot_y

    # Forward integration (x >= pivot)
    if pivot_index < len(x_vals) - 1:
        g[pivot_index + 1:] = pivot_y + np.cumsum(new_deriv[pivot_index:-1]) * dx

    # Backward integration (x < pivot)
    if pivot_index > 0:
        g[:pivot_index] = pivot_y - np.cumsum(new_deriv[1:pivot_index + 1][::-1])[::-1] * dx

    return g

def extract_sampled_modified_values_overall(x_shift=0.0, tilt=0.0):
    """
    Sample the modified CHL curve every 10 nm between 400 and 700 nm.

    Parameters:
        x_shift (float): Horizontal shift applied to the pivot.
        tilt (float): Tilt factor.

    Returns:
        np.ndarray: A 2-column matrix with wavelengths and the floored CHL values.
    """
    x_sample = np.arange(400, 701, 10)
    y_sample = CHLFit_mod_overall(x_sample, x_shift, tilt)
    y_sample_int = np.floor(y_sample)
    return np.column_stack((x_sample, y_sample_int))
```

```python
# Global variable to store the sampled modified CHL values.
global_CHLfit = None

def get_modified_CHL_array(x_shift=0.0, tilt=0.0):
    """
    Update and return the global_CHLfit array for the given x_shift and tilt.
    
    Parameters:
        x_shift (float): Horizontal shift for the CHL pivot.
        tilt (float): Tilt factor.
        
    Returns:
        np.ndarray: Updated global_CHLfit array.
    """
    global global_CHLfit
    global_CHLfit = extract_sampled_modified_values_overall(x_shift, tilt)
    return global_CHLfit
    

def plot_CHL_overall(x_shift=0.0, tilt=0.0):
    """
    Plot the modified CHL curve along with the data points and pivot markers.
    
    Parameters:
        x_shift (float): Horizontal shift for the pivot.
        tilt (float): Tilt factor.
    """
    y_vals = CHLFit_mod_overall(x_fit, x_shift, tilt)
    
    plt.figure(figsize=(8, 4))
    plt.plot(x_fit, y_vals, label='Modified CHL Curve (Opposite Tilts)', lw=2)
    plt.scatter(x_CHL, y_CHL, color='red', label='Data Points')
    plt.scatter([x_min], [y_min], color='green', s=100, label='Original Lowest Point')
    plt.scatter([x_min + x_shift], [CHLFit(x_min)], color='blue', s=100, label='Shifted Pivot')
    
    plt.xlabel('Wavelength (nm)')
    plt.ylabel('CHL Value')
    plt.title('Interactive CHL Curve with Opposite Tilts')
    plt.legend()
    plt.grid(True)
    plt.xlim(400, 700)
    plt.ylim(-100, 700)
    plt.show()

# Set up interactive widgets for CHL modification.
x_shift_slider = FloatSlider(min=-80, max=80, step=1, value=0, description='x Shift (nm)')
tilt_slider = FloatSlider(min=-0.8, max=0.8, step=0.01, value=0, description='Tilt Factor')
out = Output()

def update_display(x_shift, tilt):
    with out:
        out.clear_output(wait=True)
        plot_CHL_overall(x_shift, tilt)
        updated_array = get_modified_CHL_array(x_shift, tilt)
        print("Updated global_CHLfit array:")
        print(updated_array)

# Create the interactive interface.
interact(update_display, x_shift=x_shift_slider, tilt=tilt_slider)
display(out)
```

```python
# Function converts a wavelength (in nanometers) from the visible spectrum into an approximate RGB color value.

def wavelength_to_rgb(wavelength, gamma=0.8):
    """
    Convert a given wavelength in nm to an RGB color.
    Based on the algorithm from Dan Bruton:
    http://www.efg2.com/Lab/ScienceAndEngineering/Spectra.htm
    """
    if 380 <= wavelength < 440:
        attenuation = 0.3 + 0.7*(wavelength - 380) / (440 - 380)
        R = ((-(wavelength - 440) / (440 - 380)) * attenuation) ** gamma
        G = 0.0
        B = (1.0 * attenuation) ** gamma
    elif 440 <= wavelength < 490:
        R = 0.0
        G = ((wavelength - 440) / (490 - 440)) ** gamma
        B = 1.0 ** gamma
    elif 490 <= wavelength < 510:
        R = 0.0
        G = 1.0 ** gamma
        B = (-(wavelength - 510) / (510 - 490)) ** gamma
    elif 510 <= wavelength < 580:
        R = ((wavelength - 510) / (580 - 510)) ** gamma
        G = 1.0 ** gamma
        B = 0.0
    elif 580 <= wavelength < 645:
        R = 1.0 ** gamma
        G = (-(wavelength - 645) / (645 - 580)) ** gamma
        B = 0.0
    elif 645 <= wavelength <= 780:
        attenuation = 0.3 + 0.7*(780 - wavelength) / (780 - 645)
        R = (1.0 * attenuation) ** gamma
        G = 0.0
        B = 0.0
    else:
        R = G = B = 0.0

    return (R, G, B)
```

```python
# --- JIT-Compiled Functions for PSF and Exposure ---
@njit
def Exposure_jit(x, F):
    """
    JIT-compiled exposure function using hyperbolic tangent.
    """
    return np.tanh(F * x) / np.tanh(F)


@njit
def linear_PSF_jit(x: float, ratio: float) -> float:
    """
    JIT-compiled linear PSF function.
    """
    if ratio < 1e-6:
        return 1.0 if x >= 0 else 0.0
    if x >= ratio:
        return 1.0
    elif x <= -ratio:
        return 0.0
    else:
        return 0.5 * (1 + x / ratio)


@njit
def gaussian_PSF_jit(x: float, ratio: float) -> float:
    """
    JIT-compiled Gaussian PSF function.
    """
    if ratio < 1e-6:
        return 1.0 if x >= 0 else 0.0
    return 0.5 * (1.0 + math_erf(x / (sqrt(2) * ratio)))


@njit
def compute_edge_jit(x: float, z: float, F: float, g: float,
                     sensor_data, CHLdata, dx: float, K: float,
                     psf_mode: str) -> float:
    """
    Compute the edge response using sensor and CHL data.

    Parameters:
        x (float): x-coordinate.
        z (float): Reference position.
        F (float): Exposure parameter.
        g (float): Gamma correction factor.
        sensor_data (array-like): Data from a sensor channel.
        CHLdata (array-like): CHL data.
        dx (float): Spatial increment.
        K (float): Constant for denominator factor.
        psf_mode (str): 'gaussian' for Gaussian PSF, 'linear' for Linear PSF.

    Returns:
        float: Computed edge value.
    """
    denom_factor = sqrt(4 * K**2 - 1)
    num = 0.0
    for n in range(len(CHLdata)):
        ratio = fabs((z - CHLdata[n]) / denom_factor)
        # Select PSF based on mode.
        if psf_mode.lower() == "linear":
            weight = linear_PSF_jit(x, ratio)
        else:
            weight = gaussian_PSF_jit(x, ratio)
        num += sensor_data[n] * weight

    den = 0.0
    for n in range(len(sensor_data)):
        den += sensor_data[n]
    return Exposure_jit(num / den, F) ** g
```

```python
# --- Edge Functions Wrappers ---
def EdgeR_gaussian(x: float, z: float, F: float, g: float,
                   CHLdata, dx: float, K: float, psf_mode: str = "gaussian") -> float:
    """Edge function for the red sensor channel (Gaussian PSF)."""
    return compute_edge_jit(x, z, F, g, SensorReddata, CHLdata, dx, K, psf_mode)


def EdgeG_gaussian(x: float, z: float, F: float, g: float,
                   CHLdata, dx: float, K: float, psf_mode: str = "gaussian") -> float:
    """Edge function for the green sensor channel (Gaussian PSF)."""
    return compute_edge_jit(x, z, F, g, SensorGreendata, CHLdata, dx, K, psf_mode)


def EdgeB_gaussian(x: float, z: float, F: float, g: float,
                   CHLdata, dx: float, K: float, psf_mode: str = "gaussian") -> float:
    """Edge function for the blue sensor channel (Gaussian PSF)."""
    return compute_edge_jit(x, z, F, g, SensorBluedata, CHLdata, dx, K, psf_mode)


def EdgeR_linear(x: float, z: float, F: float, g: float,
                 CHLdata, dx: float, K: float, psf_mode: str = "gaussian") -> float:
    """Edge function for the red sensor channel (Linear PSF)."""
    return compute_edge_jit(x, z, F, g, SensorReddata, CHLdata, dx, K, psf_mode)


def EdgeG_linear(x: float, z: float, F: float, g: float,
                 CHLdata, dx: float, K: float, psf_mode: str = "gaussian") -> float:
    """Edge function for the green sensor channel (Linear PSF)."""
    return compute_edge_jit(x, z, F, g, SensorGreendata, CHLdata, dx, K, psf_mode)


def EdgeB_linear(x: float, z: float, F: float, g: float,
                 CHLdata, dx: float, K: float, psf_mode: str = "gaussian") -> float:
    """Edge function for the blue sensor channel (Linear PSF)."""
    return compute_edge_jit(x, z, F, g, SensorBluedata, CHLdata, dx, K, psf_mode)
```

```python
# Use CHLdata_0 from the global modified CHL array.
# (Ensure that global_CHLfit has been updated via get_modified_CHL_array before using.)
CHLdata_0 = global_CHLfit[:, 1] if global_CHLfit is not None else None

# --- Helper Functions for Color Fringe Calculations ---
def Farbsaum(x, z, F, g, use_ideal, CHLdata, psf_mode: str = "gaussian"):
    """
    Determine if a significant color difference exists (binary decision)
    using sensor edge responses.

    Returns:
        int: 1 if the absolute differences between channels exceed tol, else 0.
    """
    if use_ideal:
        r = EdgeR_linear(x, z, F, g, CHLdata, dx, K, psf_mode)
        g_val = EdgeG_linear(x, z, F, g, CHLdata, dx, K, psf_mode)
        b = EdgeB_linear(x, z, F, g, CHLdata, dx, K, psf_mode)
    else:
        r = EdgeR_gaussian(x, z, F, g, CHLdata, dx, K, psf_mode)
        g_val = EdgeG_gaussian(x, z, F, g, CHLdata, dx, K, psf_mode)
        b = EdgeB_gaussian(x, z, F, g, CHLdata, dx, K, psf_mode)
    return 1 if (abs(r - b) > tol or abs(r - g_val) > tol or abs(g_val - b) > tol) else 0


def Farbsaumbreite(z, F, g, use_ideal, CHLdata, psf_mode: str = "gaussian"):
    """
    Compute the overall color fringe width (binary method) for a given defocus (z).

    Returns:
        int: Aggregated binary width value.
    """
    xs = np.arange(-int(np.floor(xrange_val)), int(np.floor(xrange_val)) + 1)
    return np.sum([Farbsaum(x, z, F, g, use_ideal, CHLdata, psf_mode) for x in xs])


def ColorFringe(x, z, F, g, use_ideal, psf_mode: str = "gaussian"):
    """
    Compute the color fringe responses for a given x position.

    Returns:
        Tuple: (EdgeR, EdgeG, EdgeB) values.
    """
    if use_ideal:
        return (EdgeR_linear(x, z, F, g, CHLdata_0, dx, K, psf_mode),
                EdgeG_linear(x, z, F, g, CHLdata_0, dx, K, psf_mode),
                EdgeB_linear(x, z, F, g, CHLdata_0, dx, K, psf_mode))
    else:
        return (EdgeR_gaussian(x, z, F, g, CHLdata_0, dx, K, psf_mode),
                EdgeG_gaussian(x, z, F, g, CHLdata_0, dx, K, psf_mode),
                EdgeB_gaussian(x, z, F, g, CHLdata_0, dx, K, psf_mode))
```

```python
def evaluate_params(params):
    """
    Evaluate a single (x_shift, tilt) parameter combination by computing the 
    aggregated farbsaumbreite value over the range of defocus positions.

    For each (x_shift, tilt):
      - Compute the modified CHL curve and extract CHL data.
      - For each defocus position (z) in z_vals, calculate the Farbsaumbreite.
      - Aggregate these values using the selected aggregation_mode (mean or max).

    Parameters:
        params (tuple): (x_shift, tilt)

    Returns:
        tuple: (x_shift, tilt, aggregated_width)
    """
    x_shift, tilt = params
    # Compute the modified CHL curve for the current parameters.
    CHL_matrix = extract_sampled_modified_values_overall(x_shift, tilt)
    CHLdata_local = CHL_matrix[:, 1]
    
    # Calculate Farbsaumbreite for each defocus position.
    fringe_values = [
        Farbsaumbreite(z, F_value, 1, use_ideal=False, CHLdata=CHLdata_local, psf_mode="gaussian")
        for z in z_vals
    ]
    
    # Aggregate the fringe values based on the global aggregation_mode.
    if aggregation_mode == "mean":
        aggregated_width = np.mean(fringe_values)
    elif aggregation_mode == "max":
        aggregated_width = np.max(fringe_values)
    else:
        aggregated_width = np.mean(fringe_values)  # default fallback
    
    return (x_shift, tilt, aggregated_width)


def grid_search():
    """
    Perform a grid search over x_shift and tilt parameters using parallel threads.

    Returns:
        tuple: (best_params, results_array, x_shift_values, tilt_values)
    """
    # Define grid ranges for x_shift and tilt.
    x_shift_values = np.linspace(-80, 80, num=21)
    tilt_values = np.linspace(-0.8, 0.8, num=21)
    param_grid = [(x_shift, tilt) for x_shift in x_shift_values for tilt in tilt_values]
    
    # Evaluate each parameter combination in parallel.
    with concurrent.futures.ThreadPoolExecutor() as executor:
        results = list(executor.map(evaluate_params, param_grid))
    
    results_array = np.array(results)
    # Find the combination with the smallest aggregated fringe width.
    min_index = np.argmin(results_array[:, 2])
    best_params = results_array[min_index]
    return best_params, results_array, x_shift_values, tilt_values
```
